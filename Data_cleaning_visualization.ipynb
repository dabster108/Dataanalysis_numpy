```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn3

# Create a sample dataset
data = {
    "Category": ["A", "B", "A", "C", "B", "A", "C", "C", "B", "A"],
    "Value": np.random.randint(10, 100, 10)
}
df = pd.DataFrame(data)

# Using sets to find unique categories
unique_categories = set(df["Category"])
print("Unique Categories:", unique_categories)

# Creating two sample sets for visualization
set1 = set(df[df["Category"] == "A"]["Value"])
set2 = set(df[df["Category"] == "B"]["Value"])
set3 = set(df[df["Category"] == "C"]["Value"])

# Plot a Venn diagram for two sets
plt.figure(figsize=(5, 5))
venn2([set1, set2], ("Category A", "Category B"))
plt.title("Venn Diagram of Categories A and B")
plt.show()

# Plot a Venn diagram for three sets
plt.figure(figsize=(5, 5))
venn3([set1, set2, set3], ("Category A", "Category B", "Category C"))
plt.title("Venn Diagram of Categories A, B, and C")
plt.show()
```

```python

import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Define the cities and their coordinates
cities = ['A', 'B', 'C', 'D', 'E']
coordinates = {
    'A': (0, 0),
    'B': (2, 3),
    'C': (5, 2),
    'D': (6, 6),
    'E': (8, 3)
}

# Step 2: Create a distance matrix
def euclidean(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

distance_matrix = pd.DataFrame(index=cities, columns=cities)

for city1 in cities:
    for city2 in cities:
        if city1 != city2:
            distance_matrix.loc[city1, city2] = euclidean(coordinates[city1], coordinates[city2])
        else:
            distance_matrix.loc[city1, city2] = 0

distance_matrix = distance_matrix.astype(float)

# Step 3: Try all permutations of cities (starting and ending at the same city)
min_path = None
min_distance = float('inf')

for perm in itertools.permutations(cities[1:]):  # Fix starting at city A
    path = ['A'] + list(perm) + ['A']
    distance = 0
    for i in range(len(path) - 1):
        distance += distance_matrix.loc[path[i], path[i+1]]
    if distance < min_distance:
        min_distance = distance
        min_path = path

# Step 4: Print result
print("Shortest Path:", ' → '.join(min_path))
print("Minimum Distance: {:.2f}".format(min_distance))

# Step 5: Visualization
x_vals = [coordinates[city][0] for city in min_path]
y_vals = [coordinates[city][1] for city in min_path]

plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")

# Plot path
plt.plot(x_vals, y_vals, marker='o', linestyle='-', color='blue', markersize=10)

# Annotate cities
for city in coordinates:
    plt.text(coordinates[city][0] + 0.1, coordinates[city][1] + 0.1, city, fontsize=12)

plt.title("Shortest Path in TSP (5 Cities)")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.show()
```

```python
import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt
import seaborn as sns

# === Step 1: City coordinates and names ===
cities = ['A', 'B', 'C', 'D', 'E']
coordinates = {
    'A': (0, 0),
    'B': (2, 3),
    'C': (5, 2),
    'D': (6, 6),
    'E': (8, 3)
}

# === Step 2: Create a distance matrix ===
def euclidean(p1, p2):
    return np.round(np.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2), 2)

distance_matrix = pd.DataFrame(index=cities, columns=cities)
for i in cities:
    for j in cities:
        distance_matrix.loc[i, j] = euclidean(coordinates[i], coordinates[j])
distance_matrix = distance_matrix.astype(float)

# === Step 3: Try all permutations to find the shortest path ===
min_path = None
min_distance = float('inf')
all_paths = []

for perm in itertools.permutations(cities[1:]):  # Fix starting at 'A'
    path = ['A'] + list(perm) + ['A']
    total = 0
    for i in range(len(path)-1):
        total += distance_matrix.loc[path[i], path[i+1]]
    all_paths.append((path, total))
    if total < min_distance:
        min_distance = total
        min_path = path

# === Step 4: Print shortest path result ===
print("🔵 Shortest Path:", " → ".join(min_path))
print(f"📏 Minimum Distance: {min_distance:.2f}")
print("\n📋 Distance Matrix:\n", distance_matrix)

# === Step 5: Plot 1 - Shortest Path with labels and distances ===
def plot_path(path, title):
    x = [coordinates[city][0] for city in path]
    y = [coordinates[city][1] for city in path]

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, marker='o', linestyle='-', color='blue', linewidth=2)
    for i, city in enumerate(path):
        plt.text(x[i]+0.2, y[i]+0.2, city, fontsize=12, fontweight='bold')

    # Show distances on lines
    for i in range(len(path)-1):
        mid_x = (x[i] + x[i+1]) / 2
        mid_y = (y[i] + y[i+1]) / 2
        d = distance_matrix.loc[path[i], path[i+1]]
        plt.text(mid_x, mid_y, f"{d}", fontsize=10, color='darkred')

    plt.title(title)
    plt.xlabel("X Coordinate")
    plt.ylabel("Y Coordinate")
    plt.grid(True)
    plt.show()

plot_path(min_path, "🔵 Shortest Path with Distances")

# === Step 6: Plot 2 - Distance Matrix Heatmap ===
plt.figure(figsize=(6, 5))
sns.heatmap(distance_matrix, annot=True, cmap='Blues', fmt=".2f")
plt.title("📊 City Distance Matrix (Euclidean)")
plt.show()

# === Step 7: Plot 3 - All Possible Paths (optional preview of TSP complexity) ===
plt.figure(figsize=(8, 6))
for path, dist in all_paths:
    x = [coordinates[city][0] for city in path]
    y = [coordinates[city][1] for city in path]
    plt.plot(x, y, alpha=0.2, linestyle='--')
# Highlight the best one
plot_path(min_path, "⭐ All Paths (Faded) + Best Path (Bold)")
```
