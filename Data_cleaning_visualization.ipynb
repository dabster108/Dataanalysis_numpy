```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-10, 11) # this creates an array of integer values from -10 to 10 using NumPy's arange() function.
plt.figure(figsize=(12, 6)) # Initializes a new figure with a width of 12 inches and a height of 6 inches

plt.title('My Nice Plot')

plt.plot(x, x ** 2) #  Plots the parabola
plt.plot(x, -1 * (x ** 2)) # Plots the inverted parabola
```

```python
plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')

plt.subplot(1, 2, 1)  # rows, columns, panel selected
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])

plt.xlabel('X')
plt.ylabel('X Squared')
```

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes.plot(
    x, (x ** 2), color='red', linewidth=3,
    marker='o', markersize=8, label='X^2')

axes.plot(x, -1 * (x ** 2), 'b--', label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title("My Nice Plot")

axes.legend()

fig
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn3

# Create a sample dataset
data = {
    "Category": ["A", "B", "A", "C", "B", "A", "C", "C", "B", "A"],
    "Value": np.random.randint(10, 100, 10)
}
df = pd.DataFrame(data)

# Using sets to find unique categories
unique_categories = set(df["Category"])
print("Unique Categories:", unique_categories)

# Creating two sample sets for visualization
set1 = set(df[df["Category"] == "A"]["Value"])
set2 = set(df[df["Category"] == "B"]["Value"])
set3 = set(df[df["Category"] == "C"]["Value"])

# Plot a Venn diagram for two sets
plt.figure(figsize=(5, 5))
venn2([set1, set2], ("Category A", "Category B"))
plt.title("Venn Diagram of Categories A and B")
plt.show()

# Plot a Venn diagram for three sets
plt.figure(figsize=(5, 5))
venn3([set1, set2, set3], ("Category A", "Category B", "Category C"))
plt.title("Venn Diagram of Categories A, B, and C")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
values = np.random.randn(1000)
plt.subplots(figsize=(12, 6))

plt.hist(values, bins=100, alpha=0.8,
          histtype='bar', color='steelblue',
          edgecolor='green')
plt.xlim(xmin=-5, xmax=5)

plt.show()

```

```python
from scipy import stats

density = stats.kde.gaussian_kde(values)
density
plt.subplots(figsize=(12, 6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color='#FF7F00')
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

```python
plt.subplots(figsize=(12, 6))

plt.hist(values, bins=100, alpha=0.8, density=1,
          histtype='bar', color='steelblue',
          edgecolor='green')

plt.plot(values2, density(values2), color='#FF7F00', linewidth=3.0)
plt.xlim(xmin=-5, xmax=5)

plt.show()
```
